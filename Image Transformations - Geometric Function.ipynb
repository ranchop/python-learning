```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Stretching a curve
x = np.arange(100)
y = (x-50)**2

def stretch(x_new):
    x_old = (x_new[0]**0.5 + 50,)
    return x_old
def shape(x_old):
    return ((np.max(x_old)/2)**2,)

from scipy.ndimage import geometric_transform
y2 = geometric_transform(y,stretch,output_shape=shape(x))

fig,ax = plt.subplots(ncols = 2)
ax[0].plot(x,y)
ax[1].plot(y2)
```
