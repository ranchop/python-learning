```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
```

# Images with common colorbar

Some information

```python
# Example from https://stackoverflow.com/questions/17989917/imshow-subplots-with-the-same-colorbar 

my_image1 = np.linspace(0, 10, 10000).reshape(100,100)
my_image2 = np.sqrt(my_image1.T) + 3

fig, axes = plt.subplots(nrows=1, ncols=2)
im = axes[0].imshow(my_image1)
clim=im.properties()['clim']
axes[1].imshow(my_image2, clim=clim)

fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.5)

# plt.tight_layout()
```

```python
# Notes 
# tight_layout doesn't work if placed after fig.colorbar(...) 
# Place tight_layout before the fig.colorbar(...) command 
# x and ylabels won't be properly spaced :(  

# Generate N random images of size 
N = 4 
size = [100, 100]
images = [np.random.normal(size=size) for i in range(N)]
clim = [-3, 3]

# Setup subplots 
fig, ax = plt.subplots(ncols=N, figsize=[3*N, 3])
for i in range(N):
    im = ax[i].imshow(images[i], origin=0, clim=clim)
    ax[i].set(title='title', xlabel='xlabel', ylabel='ylabel')
plt.tight_layout()
fig.colorbar(im, ax=ax.ravel().tolist(), shrink=0.7)
# plt.tight_layout() # DO NOT use tight_layout here
```

```python
# Generate N random images of size 
N = 6
size = [100, 100]
images = [np.random.normal(size=size) for i in range(N)]
clim = [-3, 3]

# Setup subplots 
fig, ax = plt.subplots(ncols=3, nrows=2, figsize=[8, 5])
ax = ax.flatten()
for i in range(N):
    im = ax[i].imshow(images[i], origin=0, clim=clim)
    ax[i].set(title='title', xlabel='xlabel', ylabel='ylabel')
plt.tight_layout(h_pad=1, w_pad=2)
fig.colorbar(im, ax=ax.ravel().tolist(), shrink=1)
```

```python
# Generate N random images of size 
N = 3**2 
size = [100, 100]
images = [np.random.normal(size=size) for i in range(N)]
clim = [-3, 3]

# Setup subplots 
fig, ax = plt.subplots(ncols=round(N**0.5), nrows=round(N**0.5), figsize=[7, 6])
ax = ax.flatten()
for i in range(N):
    im = ax[i].imshow(images[i], origin=0, clim=clim)
    ax[i].set(title='title', xlabel='xlabel', ylabel='ylabel')
plt.tight_layout()
fig.colorbar(im, ax=ax.ravel().tolist(), shrink=1)
```
